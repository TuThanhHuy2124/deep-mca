```python
from pathlib import Path

import pandas as pd
from tqdm import trange

from deep_mca.utils import disassemble_hex
```

```python
raw_data_path = Path("../data/bhive/benchmark/throughput/skl.csv")
df = pd.read_csv(raw_data_path, header=None, names=["hex", "cycles_100"])

df["hex"] = df["hex"].fillna("").astype(str).str.strip()
df = df[df["hex"] != ""]

df["cycles_100"] = pd.to_numeric(df["cycles_100"], errors="coerce")
df = df.dropna(subset=["cycles_100"]).reset_index(drop=True)
```

```python
temp = []
for i in trange(int(df.size / 2)):
    clean_instruction_block = "\n".join(
        disassemble_hex(df.loc[i]["hex"], output_intel_syntax=False)
    )
    temp.append((clean_instruction_block, df.loc[i]["cycles_100"]))

skylake_x86 = pd.DataFrame(temp, columns=["instructions", "cycles"])
skylake_x86.to_parquet("../data/skl_x86.parquet", index=False)
```
